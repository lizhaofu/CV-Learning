```python
from PIL import Image
from matplotlib import image as mping
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
img3 = Image.open('OIP.jpeg')
plt.imshow(img3)
type(img3)
print('ok',img3.size)
```
